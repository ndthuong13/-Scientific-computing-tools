```python
# Câu 1: Chuyển đổi nhiệt độ (°C sang K)

# Nhập nhiệt độ từ người dùng
celsius = float(input("Nhập nhiệt độ (độ C): "))

# Tính nhiệt độ Kelvin
# Công thức: K = C + 273.15
kelvin = celsius + 273.15

# Xuất kết quả ra màn hình
print(f"Nhiệt độ tương ứng (K): {kelvin:.2f} K") # In với 2 chữ số thập phân
```

```python
# Câu 2: Phần nguyên và phần dư của phép chia m cho n

# Nhập hai số nguyên dương m và n (m > n)
m = int(input("Nhập số nguyên dương m: "))
n = int(input("Nhập số nguyên dương n (n < m): "))

# Tính phần nguyên (sử dụng toán tử //)
phan_nguyen = m // n

# Tính phần dư (sử dụng toán tử %)
phan_du = m % n

# In kết quả
print(f"Phần nguyên của {m} chia cho {n} là: {phan_nguyen}")
print(f"Phần dư của {m} chia cho {n} là: {phan_du}")
```

```python
# Câu 3: Tính diện tích hình tròn
import math

# Nhập bán kính từ người dùng
r = float(input("Nhập bán kính hình tròn r: "))

# Tính diện tích hình tròn (S = pi * r^2)
dien_tich = math.pi * (r ** 2)

# Xuất kết quả
print(f"Diện tích hình tròn là: {dien_tich:.2f}") # In với 2 chữ số thập phân
```

```python
# Câu 4: Tính vận tốc rơi tự do
import math

# Hằng số gia tốc trọng trường
g = 9.8 

# Nhập độ cao h (m)
h = float(input("Nhập độ cao h (mét): "))

# Tính vận tốc (v = sqrt(2 * g * h))
van_toc = math.sqrt(2 * g * h)

# Xuất kết quả với độ chính xác 2 chữ số thập phân
print(f"Vận tốc của vật khi rơi tự do là: {van_toc:.2f} m/s")
```

```python
# Câu 6: Tính giá trị biểu thức F(x, n)
import math

# Nhập x và n
x = float(input("Nhập giá trị x: "))
n = int(input("Nhập giá trị n (nguyên): "))

# Khởi tạo biến kết quả
F = 0

if n < 20: # n < 20 [cite: 14]
    # Giả định đơn giản cho trường hợp n < 20: F(x, n) = x - 2 - n
    F = x - 2 - n
    print("Áp dụng công thức F = x - 2 - n (giả định do biểu thức không đầy đủ)")
else: # Trường hợp còn lại [cite: 17]
    # Tính tổng 1 + 2 + ... + n (S = n * (n + 1) / 2) [cite: 15]
    tong = n * (n + 1) / 2
    
    # Tính mẫu số: sqrt(n + x) [cite: 16]
    mau_so = math.sqrt(n + x)
    
    # Tính F(x, n)
    if mau_so != 0:
        F = tong / mau_so
    else:
        print("Lỗi: Mẫu số bằng 0 (n + x <= 0)")
        F = None

# Xuất kết quả
if F is not None:
    print(f"Giá trị của F({x}, {n}) là: {F}")   
```

```python
# Câu 7: Kiểm tra điểm M(x, y) có nằm trong/trên hình tròn tâm I(a, b), bán kính R
import math

# Nhập tọa độ điểm M(x, y)
x = float(input("Nhập tọa độ x của M: "))
y = float(input("Nhập tọa độ y của M: "))

# Nhập tọa độ tâm I(a, b)
a = float(input("Nhập tọa độ a của I: "))
b = float(input("Nhập tọa độ b của I: "))

# Nhập bán kính R
R = float(input("Nhập bán kính R: "))

# Tính bình phương khoảng cách từ M đến I
# d^2 = (x - a)^2 + (y - b)^2
khoang_cach_binh_phuong = (x - a)**2 + (y - b)**2

# Kiểm tra điều kiện M nằm trong/trên hình tròn (d^2 <= R^2)
ket_qua = khoang_cach_binh_phuong <= (R**2)

# Xuất kết quả
print(f"Điểm M({x}, {y}) nằm trong/trên hình tròn: {ket_qua}")
```

```python
# Câu 8: Tính giá trị biểu thức A = sin(x) + cos(x) + tan(x)
import math

# Nhập giá trị x (radian)
x = float(input("Nhập giá trị x (radian): "))

# Tính các thành phần
sin_x = math.sin(x)
cos_x = math.cos(x)

# Kiểm tra tan(x) để tránh chia cho 0 (khi cos(x) = 0)
if abs(cos_x) < 1e-9: # 1e-9 là một ngưỡng nhỏ để kiểm tra gần bằng 0
    print("Giá trị tan(x) không xác định (cos(x) gần bằng 0).")
    A = None
else:
    tan_x = math.tan(x) # Hoặc tan_x = sin_x / cos_x
    A = sin_x + cos_x + tan_x
    
# Xuất kết quả
if A is not None:
    print(f"Giá trị của A là: {A:.4f}") # In với 4 chữ số thập phân
```

```python
# Câu 9: Tính tổng chuỗi B = Sum(1/k^2)
import math

# Nhập số nguyên dương n
n = int(input("Nhập số nguyên dương n: "))

B = 0.0 # Khởi tạo tổng

# Lặp từ k = 1 đến n
for k in range(1, n + 1):
    # Tính 1 / k^2
    # k_binh_phuong = pow(k, 2) # Dùng hàm pow() built-in
    k_binh_phuong = math.pow(k, 2) # Dùng hàm math.pow() 
    
    B += 1 / k_binh_phuong

# Xuất kết quả
print(f"Giá trị của B là: {B:.6f}") # In với 6 chữ số thập phân
```

```python
# Câu 10: Tính diện tích và thể tích hình cầu
import math

# Nhập bán kính r
r = float(input("Nhập bán kính hình cầu r: "))

# Tính diện tích (S = 4 * pi * r^2)
S = 4 * math.pi * (r**2)

# Tính thể tích (V = 4/3 * pi * r^3)
V = (4/3) * math.pi * (r**3)

# Xuất kết quả
print(f"Diện tích hình cầu (S): {S:.2f}")
print(f"Thể tích hình cầu (V): {V:.2f}")
```

```python
# Câu 11: Tính giá trị biểu thức C = ln(x) + e^x + sqrt(|x|)
import math

# Nhập giá trị x
x = float(input("Nhập giá trị x: "))

C = None

# logarit tự nhiên (ln) chỉ xác định với x > 0.
if x <= 0:
    print("Lỗi: ln(x) không xác định vì x <= 0.")
else:
    # logarit tự nhiên ln(x)
    ln_x = math.log(x)
    
    # hàm mũ e^x
    e_mu_x = math.exp(x)
    
    # căn bậc hai của giá trị tuyệt đối |x|
    can_tri_tuyet_doi_x = math.sqrt(abs(x))
    
    # Tính C
    C = ln_x + e_mu_x + can_tri_tuyet_doi_x
    
# Xuất kết quả
if C is not None:
    print(f"Giá trị của C là: {C:.4f}")
```

```python
# Câu 13: Tính giá trị biểu thức D(x, n)
import math

# Nhập x và n
x = float(input("Nhập giá trị x: "))
n = int(input("Nhập giá trị n (nguyên): "))

# --- Tính thành phần thứ nhất: Phân số ---
# Tử số: sin(n*x)
tu_so = math.sin(n * x)

# Mẫu số: 1 + cos^2(x)
cos_x = math.cos(x)
mau_so = 1 + (cos_x ** 2)

# Phân số
phan_so = tu_so / mau_so

# --- Tính thành phần thứ hai: Căn bậc n ---
# Căn bậc n của (x + 1)
can_bac_n = (x + 1) ** (1/n) # Sử dụng a^(1/n) để tính căn bậc n

# --- Tính D(x, n) ---
D = phan_so + can_bac_n

# Xuất kết quả
print(f"Giá trị của D({x}, {n}) là: {D:.4f}")
```
